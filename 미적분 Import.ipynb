```python
def f(x):
    return x**3 -3*x**2+x
```

```python
x = np.linspace(-1,3,400)
y = f(x)
```

```python
plt.plot(x,y)
plt.plot(0,0,'ro')
plt.plot(x,x,'r:')
plt.plot(1,-1,'go')
plt.plot(x,(3*1**2-6*1+1)*(x-1)-1, 'g--')

plt.xlim(-2,3)
plt.ylim(-4,2)
plt.xticks(np.arange(-2,4))
plt.yticks(np.arange(-1,2))

plt.annotate('',xy=(1,0), xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('',xy=(1,1), xytext=(1,0), arrowprops=dict(facecolor='gray'))

plt.annotate('', xy=(2, -1), xytext=(1, -1), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=(2, -3), xytext=(2, -1), arrowprops=dict(facecolor='gray'))

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("The slopf of f(x)")
```

```python
def relu(x):
    return np.where(x>0, x, 0)
```

```python
xx = np.linspace(-2,2,100)
plt.plot(xx,relu(xx))
```

```python
xx = np.linspace(-2,2,100)
plt.plot(xx,relu(xx))
plt.plot(0,0,marker='o',ms=10)
plt.title("ReLU")
plt.xlabel('$x$')
plt.ylabel('$ReLU(x)$')
```

```python
def f(x):
    return x**3 -3*x**2+x+6
```

```python
a,b = 0,2
```

```python
x = np.linspace(a-0.5, b+0.5,50)
```

```python
y=f(x)
```

```python
ax = plt.subplot(111)
plt.title('Integral Example')
plt.plot(x,y,'r',linewidth=2)
plt.ylim(ymin=0)

ix = np.linspace(a,b)
iy = f(ix)
verts = [(a,0)] + list(zip(ix,iy)) + [(b,0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5 * (a + b), 0.2 * (f(a) + f(b)), r"$\int_a^b f(x)dx$",
         horizontalalignment='center', fontsize=20)
plt.figtext(0.9,0.05,'$x$')
plt.figtext(0.1,0.9,'$y$')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.set_xticks((a,b))
ax.set_xticklabels(('$a$','$b$'))
ax.set_yticks([])
```
